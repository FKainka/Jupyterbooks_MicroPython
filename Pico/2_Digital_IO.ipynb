## Digital IN

```micropython
# Pin als Eingang initialisieren
from machine import Pin

button = Pin(16, Pin.IN)

print(button.value())
```

```micropython
# Überwachung des Eingangspins
import time 

while True:
    print(button.value())
    time.sleep(1)
```

```micropython
# Überwachung von Zustandsveränderungen (Extern Pullup/Pulldown)

#value = button.value()

while True:
    if button.value() != value:
        print(button.value())
        value = button.value()
        # time.sleep(1)
```

```micropython
# Interner Pulldown

button = Pin(16, Pin.IN, Pin.PULL_DOWN)

while True:
    if button.value() != value:
        print(button.value())
        value = button.value()
        # time.sleep(1)
```

```micropython
# Interner Pullup

button = Pin(16, Pin.IN, Pin.PULL_UP)

while True:
    if button.value() != value:
        print(button.value())
        value = button.value()
        # time.sleep(1)
```

## Lösungen

```micropython
# Aufgabe: Wechselblinker

from machine import Pin
import time

led1 = Pin(14, Pin.OUT)
led2 = Pin(15, Pin.OUT)

while True:
    led1.value(led2.value())
    led2.toggle()
    time.sleep(1)
```

```micropython
# Aufgabe Zwei Taster

from machine import Pin
import time

led1 = Pin(14, Pin.OUT)
led2 = Pin(15, Pin.OUT)
button1 = Pin(16, Pin.IN, Pin.PULL_DOWN)
button2 = Pin(17, Pin.IN, Pin.PULL_UP)

while True:
    if button1.value() == True:
        led1.toggle()
        time.sleep_ms(500)
        while button1.value() == True:
            time.sleep_ms(100)
    
    if button2.value() == 0:
        led2.toggle()
        while button2.value() == 0:
            time.sleep_ms(100)
```
